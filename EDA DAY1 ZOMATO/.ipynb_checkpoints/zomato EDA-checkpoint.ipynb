```python
dfc=pd.read_csv("Country-Code.csv")
```

```python
dfc
```

```python
df=pd.merge(df,dfc,on="Country Code",how="left")
```

```python
df
```

```python


country_names=df.Country.value_counts().index
df.Country.value_counts()
```

```python
country_values=df.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

Observation:Zomato has highest transactions done in INDIA (94%) followed by USA (4.73) and UK (0.87)


```python
ndf=df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'count'})
```

```python
ndf
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='count',hue='Rating color',data=ndf,palette=['black','orange','yellow','green','green'])
```

```python
sns.countplot(x='Rating color',data=ndf,palette=['white','red','orange','yellow','green','green'])
```
